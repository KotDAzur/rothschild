```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
perf_fond = pd.read_csv('/Users/paulin/Desktop/Data Rothschild/raw/perf_fonds.csv')
perf_bench = pd.read_csv('/Users/paulin/Desktop/Data Rothschild/raw/perf_bench.csv')
perf_fond['ANALYSIS_DATE_FULL']  = pd.to_datetime(perf_fond['ANALYSIS_DATE_FULL'])
perf_bench['ANALYSIS_DATE_FULL']  = pd.to_datetime(perf_bench['ANALYSIS_DATE_FULL'])
perf_bench
```

```python
perf_fond = perf_fond[perf_fond['SECURITY_NAME'].str.contains('Euro High Yield')]
perf_fond  = perf_fond[perf_fond['ANALYSIS_DATE_FULL']>="2020-07-01"]

perf_fond
```

```python
perf_bench = perf_bench[perf_bench['SECURITY_NAME'].str.contains('Euro High Yield')]
perf_bench  = perf_bench[perf_bench['ANALYSIS_DATE_FULL']>"2020-07-01"]
perf_bench
```

```python
perf_bench['SECURITY_ISIN'].value_counts()
```

```python
perf_fond['SECURITY_ISIN'].value_counts()
```

```python
perf_fond_grouped = pd.DataFrame(perf_fond.groupby('ANALYSIS_DATE_FULL')['TWR_NET_CUMULATED'].mean().reset_index())
perf_bench_grouped = pd.DataFrame(perf_bench.groupby('ANALYSIS_DATE_FULL')['TWR_NET_CUMULATED'].mean().reset_index())
```

```python
perf_bench_grouped
```

```python
perf_fond_grouped
```

```python
plt.figure(figsize=(20,6))

plt.plot(perf_fond_grouped['ANALYSIS_DATE_FULL'], perf_fond_grouped['TWR_NET_CUMULATED'])
plt.twinx()
plt.plot(perf_bench_grouped['ANALYSIS_DATE_FULL'], perf_bench_grouped['TWR_NET_CUMULATED'], color='green')
plt.title('performance du fond en fonction de la performance du benchmark')
```

```python
x = perf_fond_grouped['ANALYSIS_DATE_FULL']
y = perf_fond_grouped['TWR_NET_CUMULATED']-perf_bench_grouped['TWR_NET_CUMULATED']

plt.plot(x, y)
plt.show()
area = np.trapz(perf_fond_grouped['TWR_NET_CUMULATED']-perf_bench_grouped['TWR_NET_CUMULATED'], perf_fond_grouped['ANALYSIS_DATE_FULL'])   # Intégration par trapèzes
print(f"L'aire sous la courbe est : {area}")

# Visualisation
plt.figure(figsize=(15, 5))
plt.plot(perf_fond_grouped['ANALYSIS_DATE_FULL'], perf_fond_grouped['TWR_NET_CUMULATED']-perf_bench_grouped['TWR_NET_CUMULATED'], label="Courbe y = sin(x) + 1", color="blue")

# Remplir la zone où y >= 0
plt.fill_between(x, y, 0, where=(y >= 0), color="green", alpha=0.5, label="y >= 0 (Positif)")

# Remplir la zone où y < 0
plt.fill_between(x, y, 0, where=(y < 0), color="red", alpha=0.5, label="y < 0 (Négatif)")
plt.title("Aire sous la courbe")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
