```python
frais = pd.read_csv('data/raw/frais.csv')
filtered_frais = frais[frais['SECURITY_NAME'].isin([
    'EdR Fund Euro High Yield I - EUR',
    'EdR Fund Euro High Yield B - EUR',
    'EdR Fund Euro High Yield A - USD (H)',
    'EdR Fund Euro High Yield A - EUR',
    'EdR Fund Euro High Yield R - EUR',
    'EdR Fund Euro High Yield O - EUR',
    'EdR Fund Euro High Yield N - EUR',
    'EdR Fund Euro High Yield J - EUR',
    'EdR Fund Euro High Yield A - CHF',
    'EdR Fund Euro High Yield I - CHF (H)',
    'EdR Fund Euro High Yield K - EUR',
    'EdR Fund Euro High Yield CR - EUR',
])]
filtered_frais.DATE_FRAIS = filtered_frais.DATE_FRAIS.str.replace(' 00:00:00.000', '').str.strip()
filtered_frais.DATE_FRAIS = pd.to_datetime(filtered_frais.DATE_FRAIS, format='%Y-%m-%d')
filtered_frais['month'] = filtered_frais.DATE_FRAIS.dt.month
filtered_frais
```

```python
pivoted_frais = filtered_frais.pivot_table(index='DATE_FRAIS', columns='TYPE_FRAIS', values='MONTANT_FRAIS', aggfunc='sum').reset_index()
pivoted_frais
```

```python
pivoted_frais['year'] = pivoted_frais['DATE_FRAIS'].dt.year
pivoted_frais['month'] = pivoted_frais['DATE_FRAIS'].dt.month
pivoted_frais
```

```python
# Group by year and month, then sum each type of frais
monthly_sum_by_year = pivoted_frais.groupby(['year', 'month'])[['FRAIS_FIXE','FRAIS_ACQUIS','FRAIS_GLOBAL','FRAIS_VARIABLE']].sum().reset_index()
monthly_sum_by_year
```

```python
monthly_sum_by_year['date'] = pd.to_datetime(monthly_sum_by_year[['year', 'month']].assign(day=1)) + pd.offsets.MonthEnd(0)
monthly_sum_by_year.drop(['year', 'month'], axis=1, inplace=True)
monthly_sum_by_year
```

```python
monthly_sum_by_year.to_csv('data/processed/monthly_sum_by_year.csv', index=False)
```
