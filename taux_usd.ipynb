```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taux_usd = pd.read_excel('USGG10YR Index.xlsx')
taux_usd
```

## C'est le taux d'intérêt que le gouvernement des États-Unis paie pour emprunter de l'argent sur une période de 10 ans via des obligations.

## EN GROS TAUX DE LA BANQUE CENTRALE AMERICAINE 

# taux haut = moins d'emprunt, ralentissement de l'activité 
# taux bas = plus d'emprunts, accélératioin de l'activité

```python
plt.figure(figsize=(20,8))

plt.plot(taux_usd['Quote Date'], taux_usd['Quote Close'], color='b')
plt.xlabel('temps')  # Nom de l'axe X
plt.ylabel('taux')  # Nom de l'axe Y
plt.title('taux USD sur 10 ans')  # Titre
plt.legend()  # Légende
plt.grid()  # Grille
plt.show()
```
