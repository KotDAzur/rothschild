### travail pour comprendre les corélations possible entre volume d'un actif et son cours

```python
# on trie le df de sorte à retirer les titres de vente à découvert

pos_up = pos[pos['BALANCE_NOMINAL_OR_NUMBER']>0]
pos_up
```

```python
# combien de mesures par titres 

pos_up['SECURITY_NAME'].value_counts()
```

```python
# on sélectionne uniquement les titres sur lesquels on a au moins 30 mesures sond au moins travaillé un mois

values = pos_up['SECURITY_NAME'].value_counts()
values_to_keep = values[values >= 30].index

pos_up_30 = pos_up[pos_up['SECURITY_NAME'].isin(values_to_keep)]
pos_up_30['SECURITY_NAME'].value_counts()
```

```python
pos_up_30
```

```python
# on calcule les coefficeints de corrélation de chaque titre entre son cours et son volume
# coeff proche de 0 désigne potentiellement une opportunité manquée car pas de réaction sur le volume en fonction des varitations du cours 
# coeff proche de -1 : très bonne gestion car achat de titres quand le prix baisse et vente quand le prix monte 
# coeff > 0 : mauvaise gestion du titre, le volume augmente quand le prix augmente ou inversement, ce n'est pas ce que l'on cherche

correlations = pd.DataFrame(pos_up_30.groupby('SECURITY_NAME').apply(lambda group: group['DIRTY_PRICE'].corr(group['BALANCE_NOMINAL_OR_NUMBER'])).reset_index())

correlations = correlations.dropna()
correlations.rename(columns={0: 'correlations'}, inplace=True)
correlations.sort_values(by='correlations', inplace=True)

correlations
```

### On a réussi à trouver le coefficient de corrélation entre le volume et le cours de chaque titre dans le fond 
### on peut donc trier ces actifs en fonction des actions menées par EdR en comparant ces coefficients 

```python
selected_rows = []
last_value = -1.1  # Valeur initiale plus petite que la plus petite valeur possible

# Parcours du dataframe
for _, row in correlations.iterrows():
    current_value = row["correlations"]
    if current_value >= last_value + 0.05:
        selected_rows.append(row)
        last_value = current_value

# Création d'un nouveau dataframe avec les valeurs sélectionnées
filtered_df = pd.DataFrame(selected_rows)


filtered_df
```

```python
for titre in filtered_df['SECURITY_NAME'] : 
    plt.figure(figsize=(6,3))
    plt.plot(pos[pos['SECURITY_NAME']==titre]['INVENTORY_DATE'],pos[pos['SECURITY_NAME']==titre]['BALANCE_NOMINAL_OR_NUMBER'], color="green")
    plt.twinx()
    plt.plot(pos[pos['SECURITY_NAME']==titre]['INVENTORY_DATE'],pos[pos['SECURITY_NAME']==titre]['DIRTY_PRICE'], color="blue")
    plt.title('Evolution du volume et du cours du titre')
    plt.show()
    print("titre =",titre,", coefficient de corrélation =", filtered_df.loc[filtered_df['SECURITY_NAME'] == titre, 'correlations'].values[0])
```
