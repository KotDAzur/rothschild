```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
fixing = pd.read_excel('Fixing.xlsx')
fixing.dropna(axis=1, how='all', inplace=True)
fixing
```

```python
fixing['PRICE_DATE'] = pd.to_datetime(fixing['PRICE_DATE']).dt.date
```

```python
fixing_usd = fixing[fixing['PRICE_CURRENCY'].str.contains('USD')]
fixing_chf = fixing[fixing['PRICE_CURRENCY'].str.contains('CHF')]
fixing_gbp = fixing[fixing['PRICE_CURRENCY'].str.contains('GBP')]
fixing_usd.shape[0]+fixing_chf.shape[0]+fixing_gbp.shape[0]
```

```python
fixing_chf['PRICE_DATE'].duplicated().sum()
```

```python
plt.plot(fixing_usd['PRICE_DATE'], fixing_usd['FX_RATE'], linestyle='-', color='b')
plt.xlabel('temps')  # Nom de l'axe X
plt.ylabel('fixing')  # Nom de l'axe Y
plt.title('EUR contre USD')  # Titre
plt.legend()  # Légende
plt.grid()  # Grille
plt.show()
```

# 1 euro = Y dollar

```python
plt.plot(fixing_chf['PRICE_DATE'], fixing_chf['FX_RATE'], linestyle='-', color='b')
plt.xlabel('temps')  # Nom de l'axe X
plt.ylabel('fixing')  # Nom de l'axe Y
plt.title('EUR contre CHF')  # Titre
plt.legend()  # Légende
plt.grid()  # Grille
plt.show()
```

```python
plt.plot(fixing_gbp['PRICE_DATE'], fixing_gbp['FX_RATE'], linestyle='-', color='b')
plt.xlabel('temps')  # Nom de l'axe X
plt.ylabel('fixing')  # Nom de l'axe Y
plt.title('EUR contre CHF')  # Titre
plt.legend()  # Légende
plt.grid()  # Grille
plt.show()
```

```python
plt.plot(fixing_gbp['PRICE_DATE'], fixing_gbp['FX_RATE'], label='x^2', color='blue', linestyle='-')  # Courbe 1
plt.plot(fixing_gbp['PRICE_DATE'], fixing_usd['FX_RATE'], label='x^3', color='green', linestyle='-')  # Courbe 2
plt.plot(fixing_gbp['PRICE_DATE'], fixing_chf['FX_RATE'], label='x^4', color='red', linestyle='-')  # Courbe 3
plt.grid()
```
